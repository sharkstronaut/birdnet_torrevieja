```python
import pandas as pd
import numpy as np
import datetime
```

```python
golondrina_path = "barn_swallow_dataset/hirundo_rustica_dataset.csv"
vencejo_path = "common_swift_dataset/common_swift_dataset.csv"
```

```python
def preprocessing_data(dataset_path, file_name):
    data = pd.read_csv(dataset_path, delimiter=";")
    data = data[['OBSERVATION COUNT','COUNTY','LOCALITY','LATITUDE', 'LONGITUDE','OBSERVATION DATE']]
    data = data.loc[data['COUNTY'] == "Alicante"]
    data = data[data['LOCALITY'].str.contains("Torrevieja PNat")]
    data['OBSERVATION DATE'] = pd.to_datetime(data['OBSERVATION DATE']).apply(lambda x: x.strftime('%Y-%m'))
    # data['YEAR'] = pd.to_datetime(data['OBSERVATION DATE']).apply(lambda x: x.strftime('%Y'))
    
    data = data.loc[(data['OBSERVATION DATE'] >= '2014-01')
                     & (data['OBSERVATION DATE'] <= '2022-12')]
    data.reset_index(drop=True, inplace=True)
    data.to_csv(file_name+'.csv')
    return data
    
```

```python
golondrina_dataset = preprocessing_data(golondrina_path, "golondrina_processed")
vencejo_dataset = preprocessing_data(vencejo_path, "vencejo_processed")
```
